```python
# Import necessary packages
import PyPDF2
import re
# Open The File in the Command
file = open("fsis.pdf", 'rb')
readPDF = PyPDF2.PdfReader(file)
def find_url(string):
   #Find all the String that matches with the pattern
   regex = r"(https?://\S+)"
   url = re.findall(regex,string)
   for url in url:
      return url
# Iterating over all the pages of File
# get number of pages
#NumPages = file.getNumPages()
for page_no in range (file.getNumPages):
   page=readPDF.getPage(page_no)
   #Extract the text from the page
   text = page.extractText()
   # Print all URL
   print(find_url(text))
# CLost the file
file.close()
```
